```python
import pandas as pd
import numpy as np
import math
```

```python
dataFrame = pd.read_csv("Andhra_dataset2.csv")
```

```python
#First Column: Caste
#    1: sc
#    2: st
#    3: bc
#    4: oc
#Second Column: Age
#Third Column: Weight
#History: Toxemia Of pregnancy
#HB 
#IFA: Took folic acid.
#BP: Blood Pressure.

dataFrame.head()

```

```python
#Droping column education because education isn't really usefull anyway.
dataFrame.drop("education", inplace=True, axis=1)
```

```python
dataFrame.columns = ['Caste', 'Age', 'Weight1', 'History', 'HB', 'IFA', 'BP1', 'Resident', 'Result']
```

```python
dataFrame.head(10)
```

```python
df1 = dataFrame[dataFrame['Result'] == 0]
df2 = dataFrame[dataFrame['Result'] == 1]
```

```python
df1 = df1.fillna(df1.mean())
df2 = df2.fillna(df2.mean())
```

```python
frames = [df1, df2]
df_new = pd.concat(frames)
```

```python
df_new['Age'] = df_new['Age'].apply(np.ceil)
df_new['Weight1'] = df_new['Weight1'].apply(np.floor)
df_new['HB'] = df_new['HB'].round(1)
df_new['History'] = df_new['History'].round()
df_new['Resident'] = df_new['Resident'].round()
```
